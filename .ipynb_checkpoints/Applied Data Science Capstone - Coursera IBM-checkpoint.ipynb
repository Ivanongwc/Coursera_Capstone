## 1. Search for schools that are of walking distance to our homes

#### Let's define a query to search for schools that are within 1 kilometer from our addresses.

```python
# Define query and radius
search_query = 'School'
radius = 1000 # 1000 meters

# The url we are using
my_url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, my_latitude, my_longitude, VERSION, search_query, radius, LIMIT)
her_url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, her_latitude, her_longitude, VERSION, search_query, radius, LIMIT)
```

#### Send the GET Request and examine the results

```python
my_results = requests.get(my_url).json()
my_results
```

```python
her_results = requests.get(her_url).json()
her_results
```

#### Get relevant part of JSON and transform it into a *pandas* dataframe

```python
# assign relevant part of JSON to venues
my_venues = my_results['response']['venues']
her_venues = her_results['response']['venues']

# tranform venues into a dataframe
my_dataframe = json_normalize(my_venues)
her_dataframe = json_normalize(her_venues)
my_dataframe.head()
```

```python
her_dataframe.head()
```

#### Define information of interest and filter dataframe

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in my_dataframe.columns if col.startswith('location.')] + ['id']
my_dataframe_filtered = my_dataframe.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
my_dataframe_filtered['categories'] = my_dataframe_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
my_dataframe_filtered.columns = [column.split('.')[-1] for column in my_dataframe_filtered.columns]

my_dataframe_filtered
```

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in her_dataframe.columns if col.startswith('location.')] + ['id']
her_dataframe_filtered = her_dataframe.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
her_dataframe_filtered['categories'] = her_dataframe_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
her_dataframe_filtered.columns = [column.split('.')[-1] for column in her_dataframe_filtered.columns]

her_dataframe_filtered
```

### 2. Explore our districts

#### We are curious about the popular spots around our districts. In order to explore the area, let's start by getting the latitude and longitude values of "my place".

```python
print(my_latitude, my_longitude)
print(her_latitude, her_longitude)

my_url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, my_latitude, my_longitude, VERSION, radius, LIMIT)
her_url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, her_latitude, her_longitude, VERSION, radius, LIMIT)
```

```python
my_results = requests.get(my_url).json()
'There are {} popular areas around my area.'.format(len(my_results['response']['groups'][0]['items']))
```

```python
my_items = my_results['response']['groups'][0]['items']
my_items[0]
```

#### Process JSON and convert it to a clean dataframe

```python
my_dataframe = json_normalize(my_items) # flatten JSON

# filter columns
my_filtered_columns = ['venue.name', 'venue.categories'] + [col for col in my_dataframe.columns if col.startswith('venue.location.')] + ['venue.id']
my_dataframe_filtered = my_dataframe.loc[:, my_filtered_columns]

# filter the category for each row
my_dataframe_filtered['venue.categories'] = my_dataframe_filtered.apply(get_category_type, axis=1)

# clean columns
my_dataframe_filtered.columns = [col.split('.')[-1] for col in my_dataframe_filtered.columns]

my_dataframe_filtered.head(10)
```

#### Let's visualize these items on the map around our location

```python
my_venues_map = folium.Map(location=[my_latitude, my_longitude], zoom_start=15) # generate map centred around Ecco


# add Ecco as a red circle mark
folium.CircleMarker(
    [my_latitude, my_longitude],
    radius=10,
    popup='Her Place',
    fill=True,
    color='red',
    fill_color='red',
    fill_opacity=0.6
    ).add_to(my_venues_map)


# add popular spots to the map as blue circle markers
for lat, lng, label in zip(my_dataframe_filtered.lat, my_dataframe_filtered.lng, my_dataframe_filtered.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        fill=True,
        color='blue',
        fill_color='blue',
        fill_opacity=0.6
        ).add_to(my_venues_map)

# display map
my_venues_map
```
